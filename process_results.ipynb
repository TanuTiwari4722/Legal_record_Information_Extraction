## Library Imports

```python
import pandas as pd
import numpy as np
import pickle
import os
```

## Reading the results and converting to Dataframe

```python
output_dir = 'outputs/jsons'

all_results = []
for filename in os.listdir(output_dir):
    if filename.endswith('.pkl'):
        filepath = os.path.join(output_dir, filename)
        with open(filepath, 'rb') as file:
            data = pickle.load(file)
            data['filename'] = filename.split('.')[0]  # Extract filename without extension
        
        all_results.append(data)
```

```python
results_df = pd.DataFrame(all_results)
```
